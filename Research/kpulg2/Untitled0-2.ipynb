```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(100,10)
y = np.tan(x)
z = np.sin(x)

plt.plot(x,y,'g',x,z,'y')
plt.xlabel('Radians');
plt.ylabel('Value');
plt.title('testing')
plt.legend(['Tan','Sin'])
plt.grid()
```

```python
plt.plot(y,z)
plt.axis('equal')

plt.subplot(2,1,1)
plt.plot(y,y)
plt.title('testing')

plt.subplot(2,1,2)
plt.plot(y,z)
plt.title('tests')
```
